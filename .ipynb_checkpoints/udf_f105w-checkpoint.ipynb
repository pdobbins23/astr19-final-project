```python
import numpy as np
import sep
```

```python
from astropy.io import fits
import matplotlib.pyplot as plt
from matplotlib import rcParams

%matplotlib inline

rcParams['figure.figsize'] = [10., 8.]
```

```python
data = fits.getdata("hlsp_hudf12_hst_wfc3ir_udfmain_f105w_v1.0_drz.fits")
```

```python
m, s = np.mean(data), np.std(data)

plt.imshow(data, interpolation="nearest", cmap="gray", vmin=m-s, vmax=m+s, origin="lower")
plt.colorbar()

plt.savefig("udf_f105w_figure_1.png")
```

```python
data = data.byteswap().newbyteorder()
bkg = sep.Background(data)
```

```python
print(bkg.globalback)
print(bkg.globalrms)
```

```python
bkg_image = bkg.back()
```

```python
plt.imshow(bkg_image, interpolation="nearest", cmap="gray", origin="lower")
plt.colorbar()

plt.savefig("udf_f105w_figure_2.png")
```

```python
bkg_rms = bkg.rms()
```

```python
plt.imshow(bkg_rms, interpolation="nearest", cmap="gray", origin="lower")
plt.colorbar()

plt.savefig("udf_f105w_figure_3.png")
```

```python
data_sub = data - bkg
```

```python
objects = sep.extract(data_sub, 1.5, err=bkg.globalrms)
```

```python
len(objects)
```

```python
from matplotlib.patches import Ellipse

fig, ax = plt.subplots()
m, s = np.mean(data_sub), np.std(data_sub)

im = ax.imshow(data_sub, interpolation="nearest", cmap="gray", vmin=m-s, vmax=m+s, origin="lower")

for i in range(len(objects)):
    e = Ellipse(xy=(objects['x'][i], objects['y'][i]),
                width=6*objects['a'][i],
                height=6*objects['b'][i],
                angle=objects['theta'][i] * 180. / np.pi)
    
    e.set_facecolor("none")
    e.set_edgecolor("red")
    
    ax.add_artist(e)

fig.savefig("udf_f105w_figure_4.png")
```

```python
objects.dtype.names
```

```python
flux, fluxerr, flag = sep.sum_circle(data_sub, objects['x'], objects['y'], 3.0, err=bkg.globalrms, gain=1.0)
```

```python
for i in range(10):
    print("object {:d}: flux = {:f} +/- {:f}".format(i, flux[i], fluxerr[i]))
```
